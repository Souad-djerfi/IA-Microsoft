```python
import pandas as pd 
import matplotlib.pyplot as plt
fandango_scores=pd.read_csv("fandango_scores.csv")
fandango_scores
```

• Sélectionner les colonnes suivantes et assigner les au dataframe 
norm_reviews

```python
norm_reviews=fandango_scores[['FILM','RT_user_norm','Metacritic_user_nom','IMDB_norm','Fandango_Ratingvalue','Fandango_Stars']]
norm_reviews.head(5)
```

Créer un sous-graphique avec plt.subplot() et assigner l'objet Figure à 
la variable fig et l'objet Axes à la variable ax

```python
from numpy import arange
num_cols=['RT_user_norm','Metacritic_user_nom','IMDB_norm','Fandango_Ratingvalue','Fandango_Stars']
bar_heights= norm_reviews[num_cols].loc[1].values
bar_positions=arange(5)+1
print(bar_positions)
```

```python

fig, ax=plt.subplots()
ax.bar(bar_positions, bar_heights, width=0.5)
tick_positions=range(1,6)
ax.set_xticks(tick_positions)
ax.set_xticklabels( num_cols, rotation=90)
ax.set_xlabel("Sources de notation")
ax.set_ylabel('Note Moyenne')
ax.set_title("Moyenne des notes utilisateurs pour le film Avengers: Age of Ultron (2015)")
plt.show()




```

Diagramme en barres horizontales


```python
fig, ax=plt.subplots()
ax.barh(bar_positions, bar_heights, 0.5)
tick_positions=range(1,6)
ax.set_yticks(tick_positions)
ax.set_yticklabels( num_cols)
ax.set_ylabel("Sources de notation")
ax.set_xlabel('Note Moyenne')
ax.set_title("Moyenne des notes utilisateurs pour le film Avengers: Age of Ultron (2015)")
plt.show()

```
