```python
import numpy as np
from scipy.stats import chi2_contingency

bank_num = bank[list_num]
bank_cat = bank[list_cat]

# On dÃ©termine le tableau de contingeance
# puis on fait le test du Khi-2 pour les variables numÃ©rique
for col in list_num[:-1]:
    bank_num[col] = pd.cut(bank[col], bins=4,
                                   labels=["class1", "class2","class3", "class4", ])
                                          #"class5", "class6", "class7","class8", "class9", "class10"])

    cont = bank_num[["y", col]].pivot_table(
        index="y", columns=col, aggfunc=len).fillna(0).copy().astype(int)

    st_chi2, st_p, st_dof, st_exp = chi2_contingency(cont)
    print(col, " : ", st_p)
    

# On dÃ©termine le tableau de contingeance
# puis on fait le test du Khi-2 pour les variables catÃ©goriel    
for col in list_cat[:-1]:
    cont = bank_cat[["y", col]].pivot_table(
        index="y", columns=col, aggfunc=len).fillna(0).copy().astype(int)

    st_chi2, st_p, st_dof, st_exp = chi2_contingency(cont)
    print(col, " : ", st_p)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.model_selection import StratifiedKFold, KFold
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import accuracy_score
```

```python
num_attribs = ["age", "balance", "day", "duration", "campaign", "pdays", ]
cat_attribs = ["job", "marital", "education", "default", "housing",
               "loan",  "month", "poutcome"]

"""
"previous" et "contact" sont mise de cotÃ©
"""

num_pipeline = Pipeline([
        ('std_scaler', StandardScaler()),
    ])

full_pipeline = ColumnTransformer([
        ("num", num_pipeline, num_attribs),
        ("cat", OneHotEncoder(), cat_attribs),
    ])
```

```python
ac_score = []

#kf = KFold(n_splits=bank.shape[0]) #leave-one-out cross validation.
kf = KFold(n_splits=10) # leave-k-out cross validation
for train_index, test_index in kf.split(bank):
        
    bank_train, bank_test = bank.loc[train_index], bank.loc[test_index]
    
    X_train = full_pipeline.fit_transform(bank_train) # Attention ici on utilise la mÃ©thode fit_transforme
    y_train = np.array(bank_train["y"])

    X_test = full_pipeline.transform(bank_test)  # et ici on utilise la mÃ©thode transforme
    y_test = np.array(bank_test["y"])
    
    model = SGDClassifier(max_iter=1000, loss='log')
    model.fit(X_train, y_train)
    ac_score.append(accuracy_score(y_test, model.predict(X_test)))
    
    print(accuracy_score(y_test, model.predict(X_test)))
   
print("****************")
print("PrÃ©cision moyenne : ", np.mean(ac_score))
```

```python
ac_score = []

skf = StratifiedKFold(n_splits=10) # stratified leave-k-out cross validation

for train_index, test_index in skf.split(bank, bank['y']):
        
    bank_train, bank_test = bank.loc[train_index], bank.loc[test_index]
    
    X_train = full_pipeline.fit_transform(bank_train) # Attention ici on utilise la mÃ©thode fit_transforme
    y_train = np.array(bank_train["y"])

    X_test = full_pipeline.transform(bank_test)  # et ici on utilise la mÃ©thode transforme
    y_test = np.array(bank_test["y"])
    
    model = SGDClassifier(max_iter=1000, loss='log')
    model.fit(X_train, y_train)
    ac_score.append(accuracy_score(y_test, model.predict(X_test)))
    
    print(accuracy_score(y_test, model.predict(X_test)))

print("****************")
print("PrÃ©cision moyenne : ", np.mean(ac_score))
```

#### Mise en oeuvre de la gestion de la dyssimÃ©trie

```python
bank_yes = bank.loc[bank["y"]=="yes"]
bank_no = bank.loc[bank["y"]=="no"]

ac_score_mean = []

for i in range(10):
    
    bank_symÃ©trique = bank_yes.append(bank_no)
    bank_symÃ©trique = bank_symÃ©trique.reset_index(drop=True)
    
    bank_no = bank_no.sample(n=bank_yes.shape[0])
    bank_symÃ©trique = bank_yes.append(bank_no)
    bank_symÃ©trique = bank_symÃ©trique.reset_index(drop=True)
    
    ac_score = []
    
    skf = StratifiedKFold(n_splits=10) # stratified leave-k-out cross validation
    for train_index, test_index in skf.split(bank_symÃ©trique, bank_symÃ©trique['y']):

        bank_train, bank_test = bank_symÃ©trique.loc[train_index], bank_symÃ©trique.loc[test_index]

        X_train = full_pipeline.fit_transform(bank_train) # Attention ici on utilise la mÃ©thode fit_transforme
        y_train = np.array(bank_train["y"])

        X_test = full_pipeline.transform(bank_test)  # et ici on utilise la mÃ©thode transforme
        y_test = np.array(bank_test["y"])

        model = SGDClassifier(max_iter=1000, loss='log')
        model.fit(X_train, y_train)
        ac_score.append(accuracy_score(y_test, model.predict(X_test)))

        #print(accuracy_score(y_test, model.predict(X_test)))

    print("****************")
    print("PrÃ©cision moyenne : ", np.mean(ac_score))
    ac_score_mean.append(np.mean(ac_score))
    
print("\n####################")    
print("PrÃ©cision final : ", np.mean(ac_score_mean))
```

La prÃ©cision du modÃ¨le et semsiblement la mÃªme avec ou sans dyssimÃ©trie. On peut donc lancer l'apprantissage sur la totalitÃ© des donnÃ©es puis sauvgarder notre modÃ¨le.
